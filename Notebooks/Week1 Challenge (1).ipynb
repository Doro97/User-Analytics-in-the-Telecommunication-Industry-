```python
import warnings
warnings.filterwarnings('ignore')
pd.set_option('max_column', None)
df = pd.read_csv('Week1_challenge_data_source(CSV).csv', na_values=['?', None])
df.head()
```

```python
# the percentage of missing values in the dataset
def missing_values(x):

    # Total number of elements in the dataset
    totalCells = x.size

    #Number of missing values per column
    missingCount = x.isnull().sum()

    #Total number of missing values
    totalMissing = missingCount.sum()

    # Calculate percentage of missing values
    print("The dataset contains", round(((totalMissing/totalCells) * 100), 2), "%", "missing values.")

missing_values(df)
```

```python
# Columns with missing values 
df.isna().sum()

#percentage of missing values per column
df.isnull().mean()*100
```

```python
#drop columns with more than 50% missing values
df_clean = df.drop(['TCP DL Retrans. Vol (Bytes)','TCP UL Retrans. Vol (Bytes)','HTTP DL (Bytes)','HTTP UL (Bytes)','Nb of sec with 125000B < Vol DL','Nb of sec with 1250B < Vol UL < 6250B','Nb of sec with 31250B < Vol DL < 125000B',   
'Nb of sec with 37500B < Vol UL','Nb of sec with 6250B < Vol DL < 31250B', 'Nb of sec with 6250B < Vol UL < 37500B'], axis=1)
df_clean.head()
```

```python
#replacing missing values
clean_df=df_clean.fillna(df.mean())
clean_df.head()
```

```python
#number of xDR sessions
def num_sessions(x):
    sessions=pd.DataFrame(x.groupby('MSISDN/Number')['Bearer Id'].count())
    return sessions
sessions_df=num_sessions(clean_df)


```

```python
#duration of xDR sessions
def session_duration(x):
    duration=pd.DataFrame(x.groupby('MSISDN/Number')['Dur. (ms)'].mean())
    return duration
duration_df=session_duration(clean_df)


```

```python
#the total download (DL)
def download_data(x):
    DL_data=pd.DataFrame(x.groupby('MSISDN/Number')['Total DL (Bytes)'].sum())
    return DL_data
dl_df=download_data(clean_df)

```

```python
#the total upload (UL)
def upload_data(x):
    UL_data=pd.DataFrame(x.groupby('MSISDN/Number')['Total UL (Bytes)'].sum())
    return UL_data
ul_df=upload_data(clean_df)
```

```python
#social media applications
def sm_data(x):
    x['total_sm_data']=x['Social Media DL (Bytes)']+x['Social Media UL (Bytes)']
    total_vol=pd.DataFrame(x.groupby('MSISDN/Number')['total_sm_data'].sum())
    return total_vol
sm_df=sm_data(clean_df)
```

```python
#youtube application
def yt_data(x):
    x['total_yt_data']=x['Youtube DL (Bytes)']+x['Youtube UL (Bytes)']
    total_vol=pd.DataFrame(x.groupby('MSISDN/Number')['total_yt_data'].sum())
    return total_vol
yt_df=yt_data(clean_df)

```

```python
#netflix application
def netflix_data(x):
    x['total_netflix_data']=x['Netflix DL (Bytes)']+x['Netflix UL (Bytes)']
    total_vol=pd.DataFrame(x.groupby('MSISDN/Number')['total_netflix_data'].sum())
    return total_vol
netflix_df=netflix_data(clean_df)

```

```python
#google
def google_data(x):
    x['total_google_data']=x['Google DL (Bytes)']+x['Google UL (Bytes)']
    total_vol=pd.DataFrame(x.groupby('MSISDN/Number')['total_google_data'].sum())
    return total_vol
google_df=google_data(clean_df)
    
```

```python
#email
def email_data(x):
    x['total_email_data']=x['Email DL (Bytes)']+x['Email UL (Bytes)']
    total_vol=pd.DataFrame(x.groupby('MSISDN/Number')['total_email_data'].sum())
    return total_vol
email_df=email_data(clean_df)


```

```python
#gaming
def gaming_data(x):
    x['total_gaming_data']=x['Gaming DL (Bytes)']+x['Gaming UL (Bytes)']
    total_vol=pd.DataFrame(x.groupby('MSISDN/Number')['total_gaming_data'].sum())
    return total_vol
gaming_df=gaming_data(clean_df)
```

```python
#other applications
def other_data(x):
    x['total_other_data']=x['Other DL (Bytes)']+x['Other UL (Bytes)']
    total_vol=pd.DataFrame(x.groupby('MSISDN/Number')['total_other_data'].sum())
    return total_vol
other_data_df=other_data(clean_df)
other_data_df.head()
```

```python
#creates a dataframe containing all the new variables for each user
clean_var_df=pd.merge(sessions_df,duration_df,on='MSISDN/Number')
clean_var_df

data_df=pd.merge(dl_df,ul_df,on='MSISDN/Number')
data_df

applications_df=sm_df.merge(yt_df,on='MSISDN/Number').merge(netflix_df,on='MSISDN/Number').merge(google_df,on='MSISDN/Number')
applications_df

applications2_df=gaming_df.merge(other_data_df,on='MSISDN/Number')
applications2_df

var_df=clean_var_df.merge(data_df,on='MSISDN/Number').merge(applications_df,on='MSISDN/Number').merge(applications2_df,on='MSISDN/Number')
var_df
```

```python
#plots
import matplotlib.pyplot as plt
import seaborn as sns

def plot_hist(df:pd.DataFrame, column:str, color:str)->None:
    plt.figure(figsize=(15, 10))
    fig, ax = plt.subplots(1, figsize=(12, 7))
    #sns.displot(data=df, x=column, color=color, kde=True, height=7, aspect=2)
    #plt.title(f'Distribution of {column}', size=20)
    plt.show()

def plot_count(df:pd.DataFrame, column:str) -> None:
    plt.figure(figsize=(12, 7))
    sns.countplot(data=df, x=column)
    plt.title(f'Distribution of {column}', size=20, fontweight='bold')
    plt.show()
    

def plot_heatmap(df:pd.DataFrame, title:str, cbar=False)->None:
    plt.figure(figsize=(12, 7))
    sns.heatmap(df, annot=True, cmap='viridis', vmin=0, vmax=1, fmt='.2f', linewidths=.7, cbar=cbar )
    plt.title(title, size=18, fontweight='bold')
    plt.show()

def plot_box(df:pd.DataFrame, x_col:str, title:str) -> None:
    plt.figure(figsize=(12, 7))
    sns.boxplot(data = df, x=x_col)
    plt.title(title, size=20)
    plt.xticks(rotation=75, fontsize=14)
    plt.show()


def plot_scatter(df: pd.DataFrame, x_col: str, y_col: str, title: str, hue: str, style: str) -> None:
    plt.figure(figsize=(12, 7))
    sns.scatterplot(data = df, x=x_col, y=y_col, hue=hue, style=style)
    plt.title(title, size=20)
    plt.xticks(fontsize=14)
    plt.yticks( fontsize=14)
    plt.show()


```

```python
#PCA 

from sklearn.preprocessing import StandardScaler
features=var_df.loc[:,cols_num].values
#normalizing the features
features=StandardScaler().fit_transform(features)
features.shape
#check whether the data has been normalized
np.mean(features),np.std(features)
```

```python
feature_col=['feature'+ str(x) for x in range(features.shape[1])]
norm_data=pd.DataFrame(features,columns=feature_col)
norm_data.head()
```

```python
from sklearn.decomposition import PCA
var_pca=PCA(n_components=2)
pc_var=var_pca.fit_transform(features)

#create a dataframe that has all the pc values for all samples
pca_var_df=pd.DataFrame(data=pc_var,columns=['PC1','PC2'])
pca_var_df.head()
pca_var_df.shape
```

```python
#variance each pc holds
var_pca.explained_variance_ratio_

#PC1 holds 88.1% of the information, PC2 holds 10%
```
