```python
import warnings
warnings.filterwarnings('ignore')
pd.set_option('max_column', None)
df = pd.read_csv('Week1_challenge_data_source(CSV).csv', na_values=['?', None])
df.head()
```

```python
def num_sessions(x):
    sessions=pd.DataFrame(x.groupby('MSISDN/Number')['Bearer Id'].count())
    return sessions
num_sessions(df)


def session_duration(x):
    duration=pd.DataFrame(x.groupby('MSISDN/Number')['Dur. (ms)'].mean())
    return duration
session_duration(df)


def download_data(x):
    DL_data=pd.DataFrame(x.groupby('MSISDN/Number')['Total DL (Bytes)'].sum())
    return DL_data
download_data(df)

def upload_data(x):
    UL_data=pd.DataFrame(x.groupby('MSISDN/Number')['Total UL (Bytes)'].sum())
    return UL_data
upload_data(df)
```

```python
#social media applications
def sm_data(x):
    x['total_sm_data']=x['Social Media DL (Bytes)']+x['Social Media UL (Bytes)']
    total_vol=pd.DataFrame(x.groupby('MSISDN/Number')['total_sm_data'].sum())
    return total_vol
sm_data(df)


#youtube application
def yt_data(x):
    x['total_yt_data']=x['Youtube DL (Bytes)']+x['Youtube UL (Bytes)']
    total_vol=pd.DataFrame(x.groupby('MSISDN/Number')['total_yt_data'].sum())
    return total_vol
yt_data(df)

#netflix application
def netflix_data(x):
    x['total_netflix_data']=x['Netflix DL (Bytes)']+x['Netflix UL (Bytes)']
    total_vol=pd.DataFrame(x.groupby('MSISDN/Number')['total_netflix_data'].sum())
    return total_vol
netflix_data(df)

#google
def google_data(x):
    x['total_google_data']=x['Google DL (Bytes)']+x['Google UL (Bytes)']
    total_vol=pd.DataFrame(x.groupby('MSISDN/Number')['total_google_data'].sum())
    return total_vol
google_data(df)
    
#email
def email_data(x):
    x['total_email_data']=x['Email DL (Bytes)']+x['Email UL (Bytes)']
    total_vol=pd.DataFrame(x.groupby('MSISDN/Number')['total_email_data'].sum())
    return total_vol
email_data(df)


#gaming
def gaming_data(x):
    x['total_gaming_data']=x['Gaming DL (Bytes)']+x['Gaming UL (Bytes)']
    total_vol=pd.DataFrame(x.groupby('MSISDN/Number')['total_gaming_data'].sum())
    return total_vol
gaming_data(df)

#other
def other_data(x):
    x['total_other_data']=x['Other DL (Bytes)']+x['Other UL (Bytes)']
    total_vol=pd.DataFrame(x.groupby('MSISDN/Number')['total_other_data'].sum())
    return total_vol
other_data(df)
```

```python
#creates a dataframe containing the new variables for each user
var_df=pd.concat([num_sessions(df),session_duration(df),download_data(df),upload_data(df),sm_data(df),yt_data(df),
                  netflix_data(df),google_data(df),gaming_data(df),other_data(df)])
var_df



```

```python
# the percentage of missing values in the dataset
def missing_values(x):

    # Total number of elements in the dataset
    totalCells = x.size

    #Number of missing values per column
    missingCount = x.isnull().sum()

    #Total number of missing values
    totalMissing = missingCount.sum()

    # Calculate percentage of missing values
    print("The dataset contains", round(((totalMissing/totalCells) * 100), 2), "%", "missing values.")

missing_values(df)
```

```python
#plots
import matplotlib.pyplot as plt
import seaborn as sns

def plot_hist(df:pd.DataFrame, column:str, color:str)->None:
    plt.figure(figsize=(15, 10))
    fig, ax = plt.subplots(1, figsize=(12, 7))
    #sns.displot(data=df, x=column, color=color, kde=True, height=7, aspect=2)
    plt.title(f'Distribution of {column}', size=20)
    plt.show()

def plot_count(df:pd.DataFrame, column:str) -> None:
    plt.figure(figsize=(12, 7))
    sns.countplot(data=df, x=column)
    plt.title(f'Distribution of {column}', size=20, fontweight='bold')
    plt.show()
    
def plot_bar(df:pd.DataFrame, x_col:str, y_col:str, title:str, xlabel:str, ylabel:str)->None:
    plt.figure(figsize=(12, 7))
    sns.barplot(data = df, x=x_col, y=y_col)
    plt.title(title, size=20)
    plt.xticks(rotation=75, fontsize=14)
    plt.yticks( fontsize=14)
    plt.xlabel(xlabel, fontsize=16)
    plt.ylabel(ylabel, fontsize=16)
    plt.show()

def plot_heatmap(df:pd.DataFrame, title:str, cbar=False)->None:
    plt.figure(figsize=(12, 7))
    sns.heatmap(df, annot=True, cmap='viridis', vmin=0, vmax=1, fmt='.2f', linewidths=.7, cbar=cbar )
    plt.title(title, size=18, fontweight='bold')
    plt.show()

def plot_box(df:pd.DataFrame, x_col:str, title:str) -> None:
    plt.figure(figsize=(12, 7))
    sns.boxplot(data = df, x=x_col)
    plt.title(title, size=20)
    plt.xticks(rotation=75, fontsize=14)
    plt.show()

def plot_box_multi(df:pd.DataFrame, x_col:str, y_col:str, title:str) -> None:
    plt.figure(figsize=(12, 7))
    sns.boxplot(data = df, x=x_col, y=y_col)
    plt.title(title, size=20)
    plt.xticks(rotation=75, fontsize=14)
    plt.yticks( fontsize=14)
    plt.show()

def plot_scatter(df: pd.DataFrame, x_col: str, y_col: str, title: str, hue: str, style: str) -> None:
    plt.figure(figsize=(12, 7))
    sns.scatterplot(data = df, x=x_col, y=y_col, hue=hue, style=style)
    plt.title(title, size=20)
    plt.xticks(fontsize=14)
    plt.yticks( fontsize=14)
    plt.show()


```
